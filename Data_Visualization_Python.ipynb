## Lineplot

```python
# import libraries
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# fmri dataset
fmri = sns.load_dataset("fmri")
```

```python
fmri.head()
```

`Lineplot with timepoint and signal`

```python
sns.lineplot(x="timepoint", y="signal", data=fmri)
plt.show()
```

`Mapping another column with color`

```python
# mapping with event column
sns.lineplot(x="timepoint", y="signal", data=fmri, hue = "event")
plt.show()
```

`Adding styling for event column data`

```python
sns.lineplot(x="timepoint", y="signal", data=fmri, hue = "event", style = "event")
plt.show()
```

`Adding markers for value point`

```python
sns.lineplot(x="timepoint", y="signal", data=fmri, hue = "event", style = "event", markers = True)
plt.show()
```

## Barplot

```python
# import libraries and load pokemon dataset
import pandas as pd
pokemon = pd.read_csv('pokemon.csv')

```

```python
pokemon.head()
```

`Checking speed of legendary pokemon`

1= Legendary

0= not


```python
sns.barplot(x="is_legendary", y="speed", data=pokemon)
plt.show()
```

`Checking weight of legendary pokemon`

1= Legendary

0= not

```python
sns.barplot(x="is_legendary", y="weight_kg", data=pokemon)
plt.show()
```

`Adding hue to show pokemon generation`

```python
sns.barplot(x="is_legendary", y="speed", data=pokemon, hue="generation")
plt.show()
```

`Using pallete to asign different colour element`

```python
sns.barplot(x="is_legendary", y="speed", data=pokemon, palette="vlag")
plt.show()
```

```python
sns.barplot(x="is_legendary", y="speed", data=pokemon, color="red")
plt.show()
```

## Histogram/Dist Plot

```python
# import diamonds dataset
import pandas as pd
diamonds=pd.read_csv("diamonds.csv")
```

```python
diamonds.head()
```

`Plotting histogram based on price`

```python
sns.histplot(diamonds['price'])
plt.show()
```

`Plotting only the frequency curve`

```python
sns.distplot(diamonds['price'], hist=False)
plt.show()
```

`Adding color to distribution plot`

```python
sns.histplot(diamonds['price'], color='green')
plt.show()
```

`Plotting only the histogram`

```python
sns.histplot(diamonds['price'], kde=False, color='red')
plt.show()
```

`Changing the number of bins`

```python
sns.histplot(diamonds['price'], kde=False, color='blue', bins=10)
plt.show()
```

`Plotting on vertical axis`

```python
sns.distplot(diamonds['price'], kde=False, color='blue', bins=5, vertical=True)
plt.show()
```

## Boxplot

```python
#import telco churn dataset
churn=pd.read_csv('Churn.csv')
```

```python
churn.head()
```

`Plotting based on churn and tenure column`

```python
sns.boxplot(x='Churn', y='tenure', data=churn)
```

`Plotting based on internet service and monthly chargres`

```python
sns.boxplot(x='InternetService', y='MonthlyCharges', data=churn)
```

`Plotting based on contract`

```python
sns.boxplot(x='Contract', y='tenure', data=churn, palette='Set1')
```

`Changing line width of boxplot`

```python
sns.boxplot(x='Contract', y='tenure', data=churn, palette='Set1', linewidth=5)
```

`Plotting based on custom ordering on x axis`

```python
sns.boxplot(x='Contract', y='tenure', data=churn, palette='Set1', linewidth=3, order=['One year','Two year','Month-to-month'],hue='PaymentMethod')
```

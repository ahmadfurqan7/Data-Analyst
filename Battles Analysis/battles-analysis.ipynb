```python
# Import modul
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Load dataset
battle = pd.read_csv("battles.csv")
battle.head()
```

```python
# Check data
battle.shape
```

```python
# Rename column
battle.rename(columns={'name':'war name'},inplace=True)
battle.head()
```

```python
battle.rename(columns={'value':'point'},inplace=True)
battle.head()
```

```python
# Count source of war
battle['source'].value_counts()
```

```python
# Count target of war
battle['target'].value_counts()
```

```python
# source of war point bar plot
sns.set({'figure.figsize':(8,4)})
sns.barplot(x='source',y='point',data=battle,color='g')
plt.show()
```

```python
# target of war point bar plot
sns.set({'figure.figsize':(8,4)})
sns.barplot(x='target',y='point',data=battle,color='olive')
plt.show()
```

```python
# count plot source battles
battle['source'].value_counts()
```

```python
sns.countplot(x=battle['source'],hue=battle['source'],palette='rocket')
plt.show()
```

```python
# count plot source and target battles correlations
sns.set({'figure.figsize':(12,6)})
sns.countplot(x=battle['source'],hue=battle['target'],palette='rocket')
plt.show()
```

```python
# value of war 
plt.title('Values of war',size=30,loc='right')
sns.set({'figure.figsize':(12,8)})
sns.barplot(x='point',y='war name',data=battle,palette='Blues_d')
plt.show()
```
